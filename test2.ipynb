```python
import numpy as np
import torch
```

```python
#temp , rainfall , humidity
#one row per obsevation and one colonne per variable 

inputs = np.array([[73,67,43],
                   [91,88,64],
                   [87,134,58],
                   [102,43,37],
                   [69,96,70]], dtype= 'float32')
inputs.shape
```

```python
#targets (apples, oranges)

targets = np.array([[56,70],
                   [81,101],
                   [119,133],
                   [22,37],
                   [103,119]], dtype= 'float32')
targets.shape
```

In a linear regression model, each target variable is estimeted to be a weighted sum of
the input variables, offset by somme constant, known as a bias:

yield apple = w11 * temps + w12 * rainfall + w13 * humidity + b1              
yeild orange = w21 * temps + w22 * rainfall + w23 * humidity + b1

(w11, w12, ... ,w23 , b1 & b2) can also be presented by matrices, initialised as random values. 

```python
#weights and biases
#elements picked randomly from a normale distribution with mean 0 and standard diviation 1
w = torch.randn(2,3,requires_grad=True)
b = torch.randn(2,requires_grad=True)
print(w)
print(b)
```

# model 

```python
#our model is simply a function of the inputs and the weights(transposed) and adds the bias 
def model(x): 
    return x @ w.t() + b
```

```python
# typically we would work with tranning data: read some CSV files as numpy arrays, do some 
# processing and then convert them to pytorch tensors as follow:

inputs=torch.from_numpy(inputs)
targets=torch.from_numpy(targets)

print(inputs)
print(targets)
```

```python
#compare with target
print(targets)
```

```python
pred = model(inputs)
print(pred)
```

# MSE

```python
def mse(t1 , t2):
    diff (t1 - t2)
    return sum(diff * diff)/diff.numel()

#.numel return the number of element of diff

loss = mse(pred , targets)
print(loss)
```

# gradiant

```python
loss.backward()
print(w)
print(w.gard)
```
